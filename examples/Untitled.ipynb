```python
import pickle

with open('channel2.pickle', 'rb') as f:
    x = pickle.load(f)
```

```python
count = 0
total_length = 0
dic = []

for y in x:
    if isinstance(x[y], list):
        count += len(x[y])
        total_length = count/len(x)
for y in range(len(x)):
    dic.insert(y, list(x.values())[y][int(total_length-1)])
```
